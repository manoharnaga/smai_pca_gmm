```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram

class HierarchicalClustering:
    def linkages(self, data, linkage_type):
        linkage_matrix = linkage(data, method=linkage_type)
        return linkage_matrix

    def dendrogram(self, linkage_matrix,  labels=None):
        plt.figure(figsize=(20, 6))
        dendrogram(linkage_matrix, labels=labels, orientation='top')
        plt.title(f'Hierarchical Clustering Dendrogram')
        plt.xlabel('Data Points')
        plt.ylabel('Distance')
        plt.tight_layout()
        plt.show()
```

```python
data = pd.read_csv('SMAI-Dataset-hc-dataset/new_customers.csv')
data['Genre'] = data['Genre'].map({'Male': 0, 'Female': 1})
hc = HierarchicalClustering()
linkage_types = ['single', 'complete', 'average', 'centroid', 'ward']
for linkage_type in linkage_types:
    for feature_column in data.columns:
        subset_data = data[[feature_column]]
        linkage_matrix = hc.linkages(subset_data, linkage_type)
        hc.dendrogram(linkage_matrix, labels=None)
    linkage_matrix = hc.linkages(data, linkage_type)
    hc.dendrogram(linkage_matrix, labels=None)  
```
